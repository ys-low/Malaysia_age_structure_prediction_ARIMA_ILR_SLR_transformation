```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os 
  
# current directory 
current_dir = os.getcwd() 
relative_path=os.path.abspath(os.path.join(current_dir, os.pardir))
```

```python
test_data= pd.DataFrame(pd.read_csv(
    relative_path+"\\age_structure_data\\df_test.csv"))
test_data['Year'] = pd.to_datetime(test_data['Year'])
test_data = test_data.set_index("Year")
```

```python
auto_ori_forecast = pd.DataFrame(pd.read_csv(
    relative_path+"\\age_structure_data\\auto_base_forecast.csv"))
auto_ori_forecast['Year'] = pd.to_datetime(auto_ori_forecast ['Year'])
auto_ori_forecast =auto_ori_forecast .set_index("Year")
```
